```julia
using MatrixProductBP
using Random, Statistics
using Plots
```

```julia
N = 1
q = 4
T = 100
qs = fill(q, N)
rng = MersenneTwister(0)
A = MatrixProductBP.MPEMs.rand_mpem( [1; rand(rng, 1:7, T-2); 1], qs... );
```

```julia
overlap(x, y) = mean(x .== y);
```

```julia
sample(A)[1]
```

```julia
nsamples = 10^4
X = [sample(rng, A)[1] for _ in 1:nsamples]
Y = [sample(rng, A)[1] for _ in 1:nsamples];
```

```julia
ovl = overlap.(X,Y);
```

```julia
histogram(ovl, xlabel="overlap", normalize=true, label="")
```
